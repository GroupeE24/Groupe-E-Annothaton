```python
import requests as req
from bs4 import BeautifulSoup
url = "https://mes-questions-reponses.nathan.fr/tous-les-themes/"
response = req.get(url) #on recup une page
```

```python
import re
html = response.text
soup = BeautifulSoup(html)
```

```python
thèmes = re.findall('https://mes-questions-reponses.nathan.fr/tag/(.*)/">',html)
thèmes
```

```python
liens = []
for i in thèmes :
    liens.append('https://mes-questions-reponses.nathan.fr/tag/'+i+'/')
liens
```

## SET 1

```python
a = list(range(100))
```

```python
def premiere_phrase(html_content):
    # Utiliser BeautifulSoup pour extraire le contenu souhaité
    soup = BeautifulSoup(html_content, 'html.parser')
    descriptions = soup.find_all('span', itemprop="description")

    premieres_phrases = []
    for description in descriptions:
        # Séparation des phrases dans la description
        phrases = re.split(r'[.!?]\s+', description.text)
        if phrases:
            premieres_phrases.append(phrases[0])
    return premieres_phrases

R1 = []  # Liste pour stocker les résultats finaux

for i in a:
    if i < len(liens):  # Vérifier si l'index est dans la limite de la liste des liens
        URL = liens[i]
        reponse = req.get(URL)
        html1 = reponse.text
        # Appeler directement la fonction et ajouter le résultat à RP
        R1.extend(premiere_phrase(html1))

R1
```

```python
len(R1)
```

```python
Q1=[]
for i in a :
    url = liens[i]
    response = req.get(url)
    html = response.text
    Q1.extend(re.findall('title="(.*)" rel="bookmark">',html))
Q1=Q1[::2]
Q1
```

```python
len(Q1)
```
